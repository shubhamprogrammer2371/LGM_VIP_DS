# Importing Required Libraries

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

# Loading Or Reading The Image File

```python
Actor_Img = cv.imread("actor.jpg")
Actor_Img = cv.cvtColor(Actor_Img,cv.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.imshow(Actor_Img)
plt.axis("off")
plt.title("Actual Image")
plt.show()
```

```python
Actor_Img_Gray = cv.cvtColor(Actor_Img_Gray,cv.COLOR_BGR2GRAY)
plt.figure(figsize=(15,15))
plt.imshow(Actor_Img_Gray,cmap="gray")
plt.axis("off")
plt.title("Gray Image")
plt.show()
```

```python
Actor_Img_Invert = cv.bitwise_not(Actor_Img_Gray)
plt.figure(figsize=(15,15))
plt.imshow(Actor_Img_Invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
Actor_Img_Smooth = cv.GaussianBlur(Actor_Img_Invert,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(15,15))
plt.imshow(Actor_Img_Smooth,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
Actor_Img_Sketch = cv.divide(Actor_Img_Gray,255-Actor_Img_Smooth,scale=255)
plt.figure(figsize=(15,15))
plt.imshow(Actor_Img_Sketch,cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```
